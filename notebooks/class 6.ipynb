```python
%xmode Verbose
from jupyturtle import penup, pendown

def jump(length):
    """Move forward length units without leaving a trail.
    
    Postcondition: Leaves the pen down.
    """
    penup()
    forward(length)
    pendown()
jump(10)
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Create figure and axis
fig, ax = plt.subplots()

# Create rectangle: (x, y, width, height)
rect = patches.Rectangle((0, 0), 20, 40, 
                          linewidth=2, 
                          edgecolor='blue', 
                          facecolor='none')

# Add rectangle to plot
ax.add_patch(rect)

# Set axis limits so rectangle is visible
ax.set_xlim(-5, 30)
ax.set_ylim(-5, 50)
ax.set_aspect('equal')  # Keep proportions correct

plt.grid(True)
plt.show()
```
