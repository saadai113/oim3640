```python
def circle_area(radius):
    import math
    return math.pi * radius ** 2


def ring_area(outer, inner):
    """Area of a ring (donut shape)."""
    return circle_area(outer) - circle_area(inner)

import matplotlib.pyplot as plt
import matplotlib.patches as patches
import math

outer, inner = 5, 3

fig, ax = plt.subplots(figsize=(6, 6))

# draw outer circle
outer_circle = patches.Circle((0, 0), outer, color='#4a90d9', alpha=0.3, label=f'Outer (r={outer})')
ax.add_patch(outer_circle)

# draw ring area
ring = patches.Annulus((0, 0), outer, outer - inner, color='#7cd98a', alpha=0.5, label=f'Ring area = {ring_area(outer, inner):.2f}')
ax.add_patch(ring)

# draw inner circle
inner_circle = patches.Circle((0, 0), inner, color='#e06060', alpha=0.2, label=f'Inner (r={inner})')
ax.add_patch(inner_circle)

ax.set_xlim(-7, 7)
ax.set_ylim(-7, 7)
ax.set_aspect('equal')
ax.legend(loc='upper right')
ax.set_title('Ring Area Visualization')
ax.grid(True, alpha=0.3)
plt.show()
```

```python
def recurse(n, s):
    if n == 0:
        print(s)
    else:
        recurse(n-1, n+s)

recurse(3, 1)
import matplotlib.pyplot as plt

frames = [
    ("__main__", ""),
    ("recurse", "n=3, s=0"),
    ("recurse", "n=2, s=3"),
    ("recurse", "n=1, s=5"),
    ("recurse", "n=0, s=6  â† prints 6"),
]

fig, ax = plt.subplots(figsize=(6, len(frames) * 0.8 + 1))
for i, (name, vars_) in enumerate(reversed(frames)):
    y = i
    ax.add_patch(plt.Rectangle((0.5, y), 4, 0.7, fill=True, 
                 facecolor='#d4e6f1' if name == '__main__' else '#d5f5e3',
                 edgecolor='#333', linewidth=1.5))
    ax.text(0.7, y + 0.35, name, fontsize=12, fontweight='bold', va='center')
    ax.text(2.5, y + 0.35, vars_, fontsize=11, va='center', color='#333')

ax.set_xlim(0, 5.5)
ax.set_ylim(-0.3, len(frames) * 1.0 + 0.3)
ax.set_aspect('equal')
ax.axis('off')
ax.set_title('Stack Diagram at print(s)', fontsize=14, pad=15)
plt.tight_layout()
plt.show()
```
